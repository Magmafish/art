```python
import cv2
import matplotlib.pyplot as plt
import numpy
from skimage.io import imread
from skimage.filters import gaussian, roberts
from IPython.display import Image, display
```

```python
# Load the image
img = cv2.imread('images/HilloftheFlowers.jpg')

# Apply bilateral filter to smooth the image
smoothed = cv2.bilateralFilter(img, 9, 75, 75)

# Convert the image to grayscale
gray = cv2.cvtColor(smoothed, cv2.COLOR_BGR2GRAY)

# Apply median blur to the grayscale image
blurred = cv2.medianBlur(gray, 7)

# Apply adaptive threshold to create a binary image
thresholded = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

# Apply a color map to the binary image
painted = cv2.applyColorMap(thresholded, cv2.COLORMAP_BONE)

# Save the output image
cv2.imwrite('output.jpg', painted)
```

```python
img = cv2.imread('images/HilloftheFlowers.jpg')

# Apply a Gaussian blur to smooth out the image
blurred = cv2.GaussianBlur(img, (7, 7), 0)

# Apply the difference of Gaussians (DoG) filter to extract edges
edges = cv2.absdiff(blurred, cv2.GaussianBlur(blurred, (9, 9), 0))

# Invert the edges to create a "sketch" effect
sketch = 255 - edges

# Apply a bilateral filter to smooth out the sketch
smoothed = cv2.bilateralFilter(sketch, 15, 75, 75)

# Apply a color map to the smoothed image to create an oil painting effect
#oilpaint = cv2.applyColorMap(smoothed, cv2.COLORMAP_OIL_PAINT)

# Display the original image and the oil painting
cv2.imshow('Original Image', img)
cv2.imshow('Oil Painting', smoothed)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('output.jpg', smoothed)

cv2.imwrite('output.jpg', smoothed)
```
